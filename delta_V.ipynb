```python
Ichem = [175, 300]
Icntr = [1300, 1800]
g = 9.806
dV = np.arange(0,3e4,50)
mass = lambda Isp: 1/np.exp(dV/(g*Isp))

leo_mars = 4.3 + 0.9 + 1.4 + 4.1
mars_leo = 4.3 + 0.9 + 1.4 + 4.1
rtrip = leo_mars + mars_leo

fig = plt.figure()
ax = fig.subplots(1,1)
ax.grid(which="minor", axis="both", linewidth=0.2, alpha=0.33, zorder=-1)
ax.yaxis.set_major_formatter(mtick.PercentFormatter(xmax=1, decimals=None, symbol='%', is_latex=False))
plt.fill_between(dV/1e3, mass(Ichem[0]), mass(Ichem[1]), color="gray", label="Chemical Rockets",zorder=2)
plt.fill_between(dV/1e3, mass(Icntr[0]), mass(Icntr[1]), color="blue", label="CNTR", zorder=2)
plt.fill_between([leo_mars, leo_mars+1], [1, 1], alpha=0.3, facecolor="green", edgecolor="None",zorder=3)
plt.fill_between([rtrip, rtrip+2], [1, 1], alpha=0.3, facecolor="green", zorder=3)


plt.ylim([0,0.5])
plt.xlim([0,30])
plt.legend()
plt.ylabel("Payload / Total Mass")
plt.xlabel("$\Delta V$ (km/s)")
plt.savefig("figs\\delta_v.svg",facecolor="None")
```
